## Fillna()-- Filling the Null Values

## Filling Previous row's value

```python
df2= df.fillna(method = 'pad') 
df2
```

## Filling Next row's value

```python
df3= df.fillna(method = 'bfill') 
df3
```

## Filling Previous column's value

```python
df4= df.fillna(method = 'pad',axis= 1) 
df4
```

## Filling Next column's value

```python
df5= df.fillna(method = 'bfill',axis =1) 
df5
```

## Filling the prescribed value to the columns

```python
df6 =df.fillna({'Car':1.0,
              'BuildingArea':150.0,
              'YearBuilt':1900.0,
              'CouncilArea':'Yarra'})
df6 #Create a dictionary
```

## Filling with the Mean,Median,Min and Max value to the columns

```python
df7 = df.fillna(value =df['BuildingArea'].mean())
df7
```

```python
df8= df.fillna(value =df ['BuildingArea'].median())
df8
```

```python
df9= df.fillna(value =df ['BuildingArea'].max())
df9
```

```python
df10= df.fillna(value =df ['BuildingArea'].min())
df10
```

## Dropna() -- Drop the Null values

```python
df10 = df.dropna() #Drops all the null values
df10
```

## Droping Any and All Null Values

```python
df11 =df.dropna(how='any')
df11
```

```python
df12 =df.dropna(how='all')
df12
```
